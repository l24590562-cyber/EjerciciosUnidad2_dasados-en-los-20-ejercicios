```python
# Ejercicio 1, 20 ejercicios atravez de un menú de opciones
# Brenda Avila Gutierrez
# Ingenieria en Sistemas Computacionales
# Instituto Tecnológico de San Juan del Rio / Tecnológico Nacional de México
# Grupo: C
# Unidad 2

class FuncionesRecursivas:

    # -------- FUNCIONES RECURSIVAS --------

    def factorial(self, n):
        if n == 1:
            return 1
        else:
            return n * self.factorial(n - 1)

    def suma_natural(self, n):
        if n == 1:
            return 1
        else:
            return n + self.suma_natural(n - 1)

    def contar_digitos(self, n):
        if n < 10:
            return 1
        else:
            return 1 + self.contar_digitos(n // 10)

    def potencia(self, a, b):
        if b == 0:
            return 1
        else:
            return a * self.potencia(a, b - 1)

    def fibonacci(self, n):
        if n == 0 or n == 1:
            return n
        else:
            return self.fibonacci(n - 1) + self.fibonacci(n - 2)

    def contar_vocales(self, cadena):
        if cadena == "":
            return 0
        else:
            if cadena[0].lower() in 'aeiou':
                return 1 + self.contar_vocales(cadena[1:])
            else:
                return self.contar_vocales(cadena[1:])

    def suma_lista(self, lista):
        if not lista:
            return 0
        else:
            return lista[0] + self.suma_lista(lista[1:])

    def es_palindromo(self, palabra):
        if len(palabra) <= 1:
            return True
        elif palabra[0] != palabra[-1]:
            return False
        else:
            return self.es_palindromo(palabra[1:-1])

    def invertir_cadena(self, cadena):
        if len(cadena) == 0:
            return cadena
        else:
            return cadena[-1] + self.invertir_cadena(cadena[:-1])

    def buscar_elemento(self, lista, elemento):
        if not lista:
            return False
        elif lista[0] == elemento:
            return True
        else:
            return self.buscar_elemento(lista[1:], elemento)

    def multiplicar(self, a, b):
        if b == 0:
            return 0
        else:
            return a + self.multiplicar(a, b - 1)

    def decimal_a_binario(self, n):
        if n == 0:
            return '0'
        else:
            return self.decimal_a_binario(n // 2) + str(n % 2) if n > 1 else '1'

    def contar_caracter(self, cadena, caracter):
        if cadena == "":
            return 0
        else:
            if cadena[0] == caracter:
                return 1 + self.contar_caracter(cadena[1:], caracter)
            else:
                return self.contar_caracter(cadena[1:], caracter)

    def suma_digitos(self, n):
        if n == 0:
            return 0
        else:
            return n % 10 + self.suma_digitos(n // 10)

    def piramide(self, n):
        if n == 0:
            return
        self.piramide(n - 1)
        print('*' * n)

    def combinaciones(self, cadena, actual=""):
        if cadena == "":
            print(actual)
        else:
            self.combinaciones(cadena[1:], actual + cadena[0])
            self.combinaciones(cadena[1:], actual)

    def hanoi(self, n, origen, destino, auxiliar):
        if n == 1:
            print(f"Mover disco de {origen} a {destino}")
        else:
            self.hanoi(n - 1, origen, auxiliar, destino)
            self.hanoi(1, origen, destino, auxiliar)
            self.hanoi(n - 1, auxiliar, destino, origen)

    def es_primo(self, n, divisor=2):
        if n <= 2:
            return n == 2
        if n % divisor == 0:
            return False
        if divisor * divisor > n:
            return True
        return self.es_primo(n, divisor + 1)

    def permutaciones(self, lista, inicio=0):
        if inicio == len(lista) - 1:
            print(lista)
        else:
            for i in range(inicio, len(lista)):
                lista[inicio], lista[i] = lista[i], lista[inicio]
                self.permutaciones(lista, inicio + 1)
                lista[inicio], lista[i] = lista[i], lista[inicio]

    # -------- MENÚ PRINCIPAL --------

    def menu(self):
        while True:
            print("\n--- MENÚ DE FUNCIONES RECURSIVAS ---")
            print("1. Factorial")
            print("2. Suma Natural")
            print("3. Contar Dígitos")
            print("4. Potencia")
            print("5. Fibonacci")
            print("6. Contar Vocales")
            print("7. Suma de Lista")
            print("8. Palíndromo")
            print("9. Invertir Cadena")
            print("10. Buscar Elemento en Lista")
            print("11. Multiplicación")
            print("12. Decimal a Binario")
            print("13. Contar Caracter")
            print("14. Sumar Dígitos")
            print("15. Pirámide de Asteriscos")
            print("16. Combinaciones de una Cadena")
            print("17. Torres de Hanoi")
            print("18. Es Primo")
            print("19. Permutaciones")
            print("0. Salir")

            opcion = input("\n Seleccione una opción: \n")

            if opcion == '1':
                n = int(input("Ingrese un número: "))
                print(self.factorial(n))
            elif opcion == '2':
                n = int(input("Ingrese un número: "))
                print(self.suma_natural(n))
            elif opcion == '3':
                n = int(input("Ingrese un número: "))
                print(self.contar_digitos(n))
            elif opcion == '4':
                a = int(input("Base: "))
                b = int(input("Exponente: "))
                print(self.potencia(a, b))
            elif opcion == '5':
                n = int(input("Ingrese la posición n de la serie Fibonacci: "))
                print(self.fibonacci(n))
            elif opcion == '6':
                cadena = input("Ingrese una cadena: ")
                print(self.contar_vocales(cadena))
            elif opcion == '7':
                lista = list(map(int, input("Ingrese los números separados por espacio: ").split()))
                print(self.suma_lista(lista))
            elif opcion == '8':
                palabra = input("Ingrese una palabra: ")
                print(self.es_palindromo(palabra))
            elif opcion == '9':
                cadena = input("Ingrese una cadena: ")
                print(self.invertir_cadena(cadena))
            elif opcion == '10':
                lista = list(map(int, input("Ingrese los números separados por espacio: ").split()))
                elemento = int(input("Elemento a buscar: "))
                print(self.buscar_elemento(lista, elemento))
            elif opcion == '11':
                a = int(input("Ingrese a: "))
                b = int(input("Ingrese b: "))
                print(self.multiplicar(a, b))
            elif opcion == '12':
                n = int(input("Ingrese un número decimal: "))
                print(self.decimal_a_binario(n))
            elif opcion == '13':
                cadena = input("Ingrese una cadena: ")
                caracter = input("Caracter a contar: ")
                print(self.contar_caracter(cadena, caracter))
            elif opcion == '14':
                n = int(input("Ingrese un número: "))
                print(self.suma_digitos(n))
            elif opcion == '15':
                n = int(input("Altura de la pirámide: "))
                self.piramide(n)
            elif opcion == '16':
                cadena = input("Ingrese una cadena: ")
                self.combinaciones(cadena)
            elif opcion == '17':
                n = int(input("Cantidad de discos: "))
                self.hanoi(n, "A", "C", "B")
            elif opcion == '18':
                n = int(input("Ingrese un número: "))
                print(self.es_primo(n))
            elif opcion == '19':
                lista = list(map(int, input("Ingrese los elementos separados por espacio: ").split()))
                self.permutaciones(lista)
            elif opcion == '0':
                print("Saliendo del programa...")
                break
            else:
                print("Opción inválida. Intente de nuevo.")


# -------- EJECUCIÓN --------

if __name__ == "__main__":
    app = FuncionesRecursivas()
    app.menu()
```
